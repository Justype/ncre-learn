其实就是三个文件，把它绘制出来
- 两个声音文件：`scipy.io.wavfile.read()`
- 一个数据文件：`np.genfromtxt()`

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

rate_h, hstrain= wavfile.read("assets/H1_Strain.wav", "rb")
rate_l, lstrain= wavfile.read("assets/L1_Strain.wav", "rb")
reftime, ref_H1 = np.genfromtxt('assets/wf_template.txt').transpose()
# time data ？

htime_interval = 1/rate_h # 得到波形的时间间隔
ltime_interval = 1/rate_l
fig = plt.figure(figsize=(12, 6)) # 创建绘图空间

# 丢失信号起始点
# region H1_Strain
#           矩阵第一维度的长度 / 比例？= 函数在坐标轴上的总长度
htime_len = hstrain.shape[0]/rate_h
htime = np.arange(-htime_len/2, htime_len/2 , htime_interval) # 创建时间序列
plth = fig.add_subplot(221) # 划分好空间
plth.plot(htime, hstrain, 'y') # 划线
plth.set_xlabel('Time (seconds)')
plth.set_ylabel('H1 Strain')
plth.set_title('H1 Strain')
# endregion

# region H1_Strain
ltime_len = lstrain.shape[0]/rate_l
ltime = np.arange(-ltime_len/2, ltime_len/2 , ltime_interval)
pltl = fig.add_subplot(222)
pltl.plot(ltime, lstrain, 'g')
pltl.set_xlabel('Time (seconds)')
pltl.set_ylabel('L1 Strain')
pltl.set_title('L1 Strain')
# endregion

pltref = fig.add_subplot(212)
pltref.plot(reftime, ref_H1)
pltref.set_xlabel('Time (seconds)')
pltref.set_ylabel('Template Strain')
pltref.set_title('Template')
fig.tight_layout()

# plt.savefig("Gravitational_Waves_Original.png")
plt.show()
plt.close(fig)
```
