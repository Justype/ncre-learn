# Matplotlib

- Matplotlib库由各种可视化类构成，内部结构复杂，受Matlab启发
- matplotlib.pyplot是绘制各类可视化图形的命令子库，相当于快捷方式
    - `import matplotlib.pyplot as plt`

## 简单的使用

```python
import matplotlib.pyplot as plt

ls = [3, 1, 4, 5, 2]
plt.plot(ls) # 就是一个简单的列表
plt.ylabel("Grade") # 设置y轴标签  （从名字就看出来了）
                    # 中文显示的不好
# plt.savefig("test", dpi = 400) # 保存文件，默认PNG
plt.show()
```

列表索引为横坐标，值为纵坐标

### 传入 (x, y) 两个坐标

```python
import matplotlib.pyplot as plt

x = [0, 2, 3, 6, 9]
plt.plot(x, ls) # 传入两个列表
plt.ylabel("Grade")
plt.axis((-1, 10, 0, 6)) # 设置坐标系的范围，先x后y
plt.show()
```

## 绘图区域

- plt.subplot(nrows, ncols, plot_number) 可简写为一个数
- 行，列，选择第一个区域（左上为1，左->右，上->下）

在全局绘图区域中创建一个分区体系，并定位到一个子绘图区域

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    '''
    能量衰减函数
    '''
    return np.exp(-t) * np.cos(2 * np.pi * t)

a = np.arange(0, 5, 0.02)

plt.subplot(211)
plt.plot(a , f(a))

plt.subplot(2, 1, 2)
plt.plot(a, np.cos(2 * np.pi * a), 'r--') # 绘制正弦曲线，格式设置为曲线

plt.show()
```

## plot 函数

`plt.plot(x, y, format_string, **kwargs)`
- x : X轴数据，列表或数组，可选
- y : Y轴数据，列表或数组
- format_string: 控制曲线的格式字符串，可选
- **kwargs : 第二组或更多(x,y,format_string)
    - 当绘制多条曲线时，各条曲线的x不能省略

多条曲线的例子

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(10)
plt.plot(a, a, a, 2*a, a, 3*a, a, 4*a)
plt.show()
```

### format_string

颜色字符 + 标记字符 + 风格字符
- 颜色字符
    - 英文首字母：r(ed), b(lue), g(reen), y(ellow)
        - (blac)k, w(hite), m(agenta), c(yan)
    - RGB：`#FFFFFF`
    - 灰度值 `0-1`
- 风格字符
    - '‐' 实线
    - '‐‐' 破折线
    - '‐.' 点划线
    - ':' 虚线
    - '' ' ' 无线条


```python
plt.plot(a, a, "-", a, 2*a, "--", a, 3*a, "-.", a, 4*a, ":")
plt.show()
```

### 1 rcParams

属性          | 说明
--            | --
'font.family' | 用于显示字体的名字
'font.style'  | 字体风格，正常'normal'或 斜体'italic'
'font.size'   | 字体大小，整数字号或者'large'、'x‐small'

恢复默认设置 `.rcdefaults()`

plt 下也有 rcParams，二者是等价的，不用再多引用 matplotlib 吧

```python
import matplotlib.pyplot as plt
import numpy as np

matplotlib.rcParams["font.family"] = "Microsoft YaHei"
plt.rcParams["font.size"] = 20

x = np.arange(0, 5, 0.02)
y = np.cos(2 * np.pi * x)

plt.ylabel("振幅")
plt.xlabel("时间")
plt.plot(x, y, "--")

plt.show()
plt.rcdefaults() # 恢复默认设置
```

### fontproperties （建议）

文字显示的函数，都有该参数

```python
plt.ylabel("振幅", fontproperties="Microsoft YaHei", fontsize=20)
plt.xlabel("时间", fontproperties="Microsoft YaHei", fontsize=20)
plt.plot(x, y, "r--")
plt.show()
```

## 文字显示

常用函数

函数           | 说明
--             | --
plt.xlabel()   | 对X轴增加文本标签
plt.ylabel()   | 对Y轴增加文本标签
plt.title()    | 对图形整体增加文本标签
plt.text()     | 在任意位置增加文本
plt.annotate() | 在图形中增加带箭头的注解

支持 LaTex，与 Jupyter 一样，在`$ $`之中

`plt.annotate(s, xy=arrow_crd, xytext=text_crd, arrowprops=dict)`
- s：字符串
- xy：箭头位置
- xytext：文本位置

直接看文档不就好了

https://matplotlib.org/3.3.2/api/text_api.html#matplotlib.text.Annotation

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 5, 0.02)
y = np.cos(2 * np.pi * x)
plt.plot(x, y, "--")

plt.ylabel("振幅", fontproperties="Microsoft YaHei", fontsize=15, color="green")
plt.xlabel("时间", fontproperties="Microsoft YaHei", fontsize=15)

plt.title(r"正弦波 $y=cos(2\pi x)$", fontproperties="Microsoft YaHei", fontsize=25)
plt.text(4.5, 1.05, "some \n text", fontsize=15)

plt.annotate("$x=2$", xy=(2,1), xytext=(2.5, 1.25), arrowprops={
    "facecolor" : "black",
    "shrink" : 0.1,
    "width" : 1,
    "headwidth" : 6,
    "headlength" : 4
})

plt.annotate("$x=0,y=1$", (0,1), xytext=(.5, 1.25),
    arrowprops = dict(arrowstyle="->"))

plt.axis((-0.5, 5.5, -1.5, 1.5))
plt.grid(True)

plt.show()
```

## 子绘图区域

其实这和WPF等绘制UI的Grid没有区别

`plt.subplot2grid(GridSpec, CurSpec, colspan=1, rowspan=1)`
- 设定网格 (行, 列)
- 选中网格
- 确定选中行列区域数量，编号从0开始

```python
import matplotlib.pyplot as plt

def text_and_remove_xy(s: str, color):
    """
    去除 x,y 刻度，在中间展示一段文本
    """
    plt.text(0.5, 0.5, s, color=color, 
        verticalalignment="center", 
        horizontalalignment="center")
    plt.xticks([]) # 去除x轴刻度
    plt.yticks([]) # 去除y轴刻度

plt.subplot2grid((3,3), (0,0), colspan=2, rowspan=2)
text_and_remove_xy("ax1", "black")

plt.subplot2grid((3,3), (0,2))
text_and_remove_xy("ax2", "violet")

plt.subplot2grid((3,3), (1,2), rowspan=2)
text_and_remove_xy("ax3", "green")

plt.subplot2grid((3,3), (2,0))
text_and_remove_xy("ax4", "crimson")

plt.subplot2grid((3,3), (2,1))
text_and_remove_xy("ax5", "purple")

plt.show()
```

### GridSpec 类

用于子区域设计

`import matplotlib.gridspec`



```python
import matplotlib.gridspec

gs = matplotlib.gridspec.GridSpec(3,3)

plt.subplot(gs[:2,:2])
text_and_remove_xy("ax1", "black")

plt.subplot(gs[0,-1])
text_and_remove_xy("ax2", "violet")

plt.subplot(gs[1:,2])
text_and_remove_xy("ax3", "green")

plt.subplot(gs[2,0])
text_and_remove_xy("ax4", "crimson")

plt.subplot(gs[2,1])
text_and_remove_xy("ax5", "purple")
```
