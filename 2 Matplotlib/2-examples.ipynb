## Pie Chart

[pie api](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html?highlight=pie#matplotlib.pyplot.pie)

```python
import matplotlib.pyplot as plt

plt.pie(
    x=(15, 30, 45, 10),
    explode=(0, 0.1, 0, 0.2),
    labels=("Frogs", "Hogs", "Dogs", "Logs"),
    colors=("orange", "azure", "violet", "lightblue"),
    autopct="%1.1f%%",
    shadow=True,
    startangle=90,
)

plt.show()
```

## Histogram

[histogram api](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html#matplotlib.pyplot.hist)

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.random.normal( # a 为正态分布数组
    loc=100,  # 均值
    scale=20,  # 标准差
    size=200
)

plt.hist(
    x=a,
    bins=20, # 直方图个数
    histtype="stepfilled",
    facecolor="crimson",
    alpha=0.75,
    density=True, # 这两项设为 True，
    stacked=True, # 总和为1
    # orientation="horizontal",
    rwidth=0.2
)

plt.show()
```

## 极坐标

这里使用面对对象的方法

是这个[projections.polar](https://matplotlib.org/stable/api/projections_api.html#module-matplotlib.projections.polar)吗

使用的是[axes](https://matplotlib.org/stable/api/axes_api.html)

```python
import numpy as np
import matplotlib.pyplot as plt

N = 20
theta = np.linspace(0, 2*np.pi, N, endpoint=False) # 生成 0-2pi 的等间距数组
radii = 10 * np.random.rand(N) # 生成半径随机数
width = np.pi / 4 * np.random.rand(N) # 生成角度随机数

ax = plt.subplot(projection="polar")
bars = ax.bar(theta, radii, width=width, bottom=0) # 返回柱对象集合

for r,bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.viridis(r / 10))
    bar.set_alpha(0.5) # 透明度？

plt.show()
```

## 散点图
